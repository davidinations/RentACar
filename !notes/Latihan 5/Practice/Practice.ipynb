```python
def addToInventory(inventory, addedItems):
    # your code goes here
    for i in addedItems:
        if i in inventory.keys():
            inventory[i] += 1
        else:
            inventory[i] = 1
    # convert dict to string
    print('Inventory:')
    total = 0
    for k, v in inventory.items():
        print(str(v) + ' ' + k)
        total += v
    print('\nTotal number of items: ' + str(total))  

inventory = {'gold coin': 42, 'rope': 1}
dragonLoot = ['gold coin', 'dagger', 'gold coin', 'gold coin', 'ruby']
addToInventory(inventory, dragonLoot)
```

```python
def first_repeating_character(text):
    """
    This function finds the first character that indicates a repeat in a string.

    Args:
      text: The string to check.

    Returns:
      The first character that indicates a repeat, or None if there are no repeating characters.
    """
    seen_characters = set()
    for character in text:
      if character in seen_characters:
        return character
      else:
        seen_characters.add(character)
    return None

def first_non_repeating_character(text):
    """
    This function finds the first non-repeating character in a string.

    Args:
      text: The string to check.

    Returns:
      The first non-repeating character, or None if there are no non-repeating characters.
    """
    character_counts = {}
    for character in text:
      if character not in character_counts:
        character_counts[character] = 1
      else:
        character_counts[character] += 1

    for character, count in character_counts.items():
      if count == 1:
        return character

    return None

def repeatingCount(text):
    print(f'Non-repeating character : {first_non_repeating_character(text)}')
    print(f'Repeating character : {first_repeating_character(text)}')

text = "123PurwadhikaHeroesHeroes"
repeatingCount(text)
```

```python
def cekHeterogram(text):
    #define what word is (if there is 2 word or more split it into list)
    text = text.split(' ')
    #define if there is same word at 1 word
    for i in text:
        if len(i) != len(set(i)):
            return False
        else:
            return True

assert cekHeterogram('purwadhika jogja') == False
assert cekHeterogram('jcds 0210') == True
assert cekHeterogram('  ') == True
```

```python
def convertInteger(num):
    if num == 0:
        return '0'

    is_negative = False
    if num < 0:
        is_negative = True
        num = -num

    result = []
    while num > 0:
        digit = num % 10
        result.append(chr(ord('0') + digit))
        num //= 10

    if is_negative:
        result.append('-')

    result.reverse()
    return ''.join(result)

assert convertInteger(0) == '0'
assert convertInteger(150) == '150'
assert convertInteger(1935) == '1935'
assert convertInteger(-1234) == '-1234'
```

```python
def modeValue(list1):
    if len(list1) == 0:
        return None
    else:
        if len(list1) == len(set(list1)):
            return None
        else:
            return max(set(list1), key = list1.count)

assert modeValue([]) == None
assert modeValue([1, 2, 3, 4, 4]) == 4
assert modeValue([1, 1, 2, 3, 4]) == 1
assert modeValue([1.5, 1.4, 2.1, 3.3, 4.9]) == None
```
