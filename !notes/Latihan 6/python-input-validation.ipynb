 <h1><p style='text-align: center; font-weight: bold;'> Input Validation </p></h1>

Validasi input memeriksa bahwa nilai yang dimasukkan oleh pengguna, seperti teks dari fungsi input(), diformat dengan benar. Sebagai contoh, jika Anda ingin pengguna memasukkan usia, kode Anda tidak boleh menerima jawaban yang tidak masuk akal seperti angka negatif (yang berada di luar rentang bilangan bulat yang dapat diterima) atau kata-kata (yang merupakan tipe data yang salah). **Validasi input dapat mencegah bug atau kerentanan keamanan**.

___

**Menulis kode validasi input untuk setiap pemanggilan input() dalam program Anda akan menjadi membosankan. Selain itu, Anda mungkin melewatkan beberapa kasus tertentu dan membiarkan input yang tidak valid melewati pemeriksaan Anda**. Pada bab ini, Anda akan belajar bagaimana menggunakan modul PyInputPlus pihak ketiga untuk validasi input.

PyInputPlus memiliki beberapa fungsi untuk berbagai jenis input:

* `inputStr()` Sama seperti fungsi input() bawaan tetapi memiliki fitur PyInputPlus secara umum. Anda juga dapat mengoper fungsi validasi kustom ke fungsi ini.

* `inputNum()` Memastikan pengguna memasukkan angka dan mengembalikan sebuah int atau float, tergantung pada apakah angka tersebut memiliki titik desimal di dalamnya.

* `inputChoice()` Memastikan pengguna memasukkan salah satu pilihan yang disediakan.

* `inputMenu()` Mirip dengan `inputChoice()`, tetapi menyediakan menu dengan pilihan bernomor atau huruf.

* `inputDatetime()` Memastikan pengguna memasukkan tanggal dan waktu.

* `inputYesNo()` Memastikan pengguna memasukkan respons "ya" atau "tidak".

* `inputBool()` Mirip dengan `inputYesNo()`, tetapi mengambil respons "True" atau "False" dan mengembalikan nilai Boolean.

* `inputEmail()` Memastikan pengguna memasukkan alamat email yang valid.

* `inputFilepath()` Memastikan pengguna memasukkan jalur file dan nama file yang valid, dan secara opsional dapat memeriksa apakah file dengan nama tersebut ada.

* `inputPassword()` Sama seperti input() bawaan, tetapi menampilkan karakter * saat pengguna mengetik sehingga kata sandi, atau informasi sensitif lainnya, tidak ditampilkan di layar.

Reference: https://pyinputplus.readthedocs.io/en/latest/

Parameter berikut ini tersedia untuk semua fungsi input*(). Anda dapat melihat dokumentasi ini dengan memanggil `help(pyip.parameters)`:

```python
# Importing module
import pyinputplus as pyip
help(pyip.parameters)
```

PyInputPlus akan terus meminta teks kepada pengguna sampai mereka memasukkan input yang valid. Disarankan untuk mengimpor PyInputPlus dengan nama yang lebih pendek, yaitu pyip. Berikut beberapa contoh penggunaan metode yang ada di dalam modul:

```python
# Contoh penggunaan metode inputMenu()
prompt = f"""
Selamat datang di pasar buah
List menu:

"""
choice = ['Show', 'Add', 'Delete', 'Buy', 'Exit']
response = pyip.inputMenu(prompt=prompt, choices=choice, numbered=True)
print(response)
```

```python
# Contoh penggunaan metode inputYesNo()
response = pyip.inputYesNo(prompt='Apakah anda ingin keluar(yes/no): ')
print(response)
```

```python
# Contoh penggunaan metode inputInt()
response = pyip.inputInt(prompt='Masukkan jumlah buah: ', greaterThan=5)
print(response)
```

```python
# Contoh penggunaan metode inputStr()
prompt = "Masukkan nama buah: "
response = pyip.inputStr(prompt=prompt, applyFunc=lambda x: x.capitalize(), blockRegexes=[r'[0-9]'])
print(response)
```
