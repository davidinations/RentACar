### **1. class statement**

Kita pernah mengisi formulir kertas atau elektronik seperti untuk kunjungan dokter, pembelian online, atau untuk melakukan RSVP ke pesta pernikahan. Formulir ada sebagai cara yang seragam untuk mengumpulkan informasi dari orang lain. Formulir yang berbeda berisi jenis informasi yang berbeda, misalnya formulir dokter melaporkan kondisi medis pasien, sedangkan RSVP pernikahan akan melaporkan setiap tamu pernikahan, tidak sebaliknya.

Dalam Python, seperti halnya kertas atau formulir elektronik, class adalah cetak biru untuk objek Python, yang berisi data yang merepresentasikan kata "benda". Kata benda ini bisa berupa pasien dokter, pembelian e-commerce, atau tamu pernikahan. Class seperti template formulir kosong, dan objek seperti formulir yang sudah diisi dengan data aktual tentang jenis hal yang direpresentasikan oleh formulir tersebut.

```python
import random

class rsvpForm:
    def __init__(self, name, accepted, guest):
        self.name = name
        self.has_accepted = accepted
        self.n_guest = guest
    
    def confirm(self):
        if self.has_accepted == True:
            self.invitation_num = random.randint(1, 100) 
            print(f'Terima kasih sudah datang!')
        else:
            self.invitation_num = 0
            self.n_guest = 0
            print('Semoga kita bertemu di lain kesempatan!')     
```

Program di atas mendefinisikan sebuah class bernama rsvpForm dengan statement class, di dalamnya terdapat definisi untuk dua metode yaitu `__init__()` dan `confirm()`. Suatu class tidak diharuskan memiliki metode `__init__()`, tetapi hampir selalu ada. Metode `__init__()` adalah tempat Anda biasanya mengatur nilai awal atribut dan tiak pernah memiliki pernyataan return.

Pemanggilan rsvpForm() akan membuat objek rsvpForm dan menjalankan kode dalam metode `__init__()` untuk objek tersebut. Kita memberikan tiga buah value sebagai argumen rsvpForm() yang akan diteruskan ke parameter `__init__()`. Argumen-argumen ini disimpan ke atribut **self.name, self.has_accepted, dan self.n_guest**. Selain itu, kita dapat memanggil metode `confirm()` pada objek rsvpForm. Metode ini akan membangkitkan secara random satu nomor undangan yang akan di simpan dalam atribut **self.invitation_num**, dan memperbarui nilai **self.n_guest**.

Misalnya, kita coba memanggil objek rsvpForm() kemudian kita tetapkan pada variabel **obj1** dan **obj2**. Setiap objek memiliki kumpulan atributnya sendiri. Anda dapat menganggap atribut objek mirip dengan key pada dictionary, artinya Anda dapat mengakses dan mengatur atribut-atribut ini seperti membaca, memodifikasi nilai, dan memberikan atribut baru pada objek tersebut. Secara teknis, metode juga dianggap sebagai atribut dari sebuah kelas.

```python
obj1 = rsvpForm('Nabil', True, 2)
obj1.confirm()

print(f'Name: {obj1.name}')
print(f'Guest: {obj1.n_guest}')
print(f'Invitation number: {obj1.invitation_num}')
```

```python
obj2 = rsvpForm('Andi', False, 1)
obj2.confirm()

print(f'Name: {obj2.name}')
print(f'Guest: {obj2.n_guest}')
print(f'Invitation number: {obj2.invitation_num}')
```

### **2. Inheritance**

Mirip dengan fungsi, inheritance adalah teknik penggunaan ulang kode yang dapat Anda terapkan pada class. Ini adalah tindakan menempatkan class ke dalam hubungan parent-child di mana child class mewarisi salinan metode parent class, dan membebaskan Anda dari menduplikasi metode di beberapa class. Banyak programmer menganggap inheritance terlalu berlebihan atau bahkan berbahaya karena kompleksitas dalam sebuah program. Tetapi penggunaan teknik ini secara terbatas dapat menghemat banyak waktu dalam mengatur kode dengan lebih clean.

```python
class ParentClass:
    def method_A(self):
        print('Hello, world!')

class ChildClass(ParentClass): 
    def method_B(self):
        print("ParentClass objects tidak memiliki method B.")
                    
class GrandchildClass(ChildClass): 
    def method_C(self):
        print('Hanya GrandchildClass objects yang memiliki method C.')

print('ParentClass object :')
parent = ParentClass()
parent.method_A()

print('\nChildClass object :')
child = ChildClass()
child.method_A()
child.method_B()

print('\nGrandchildClass object :')
grandchild = GrandchildClass()
grandchild.method_A()
grandchild.method_B()
grandchild.method_C()
```
